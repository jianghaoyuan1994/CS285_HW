```python
list_dir = os.listdir()
```

```python
list_dir.remove('dqn_test_pong_PongNoFrameskip-v4_28-04-2020_23-55-47')
list_dir.remove('dqn_test_pong_LunarLander-v2_29-04-2020_21-40-27')
list_dir.remove('dqn_test_pong_LunarLander-v2_29-04-2020_09-00-55')
list_dir.remove('dqn_test_pong_LunarLander-v2_29-04-2020_08-40-32')
list_dir.remove('.ipynb_checkpoints')
list_dir.remove('Untitled.ipynb')
```

```python
for i in list_dir:
    path = os.path.join(i, file_name)
    df = pd.read_csv(path)
    step = df["Step"]
    value = df["Value"]
    plt.plot(step, value, label=i[7:30])
    plt.legend()
```
